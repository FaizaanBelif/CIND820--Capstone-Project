```python
#Load the required libraries
import pandas as pd
import numpy as np
import seaborn as sns

#Load the data
df = pd.read_csv('bankfull.txt',sep=';')


#View the data
df.head(10)
```

```python
#Basic information

df.info()

#Describe the data

df.describe()
```

```python
#Find the duplicates

df.duplicated().sum()
```

```python
#unique values

df['age'].unique()

#df['marital'].unique()

#df['job'].unique()

#df['education'].unique()
```

```python
#Plot the unique values

sns.countplot(x=df["age"]).unique()
```

```python
#Find null values

df.isnull().sum()
```

```python
#Replace null values
#As there is no null values, the data remains the same

df.replace(np.nan,'0',inplace = True)

#Check the changes now
df.isnull().sum()
```
